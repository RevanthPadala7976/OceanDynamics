# Ocean Acidification

Ocean acidification, often called the 'evil twin' of global warming, seriously threatens our planet's oceans. It occurs when the ocean absorbs excess carbon dioxide from the atmosphere, which leads to chemical reactions that reduce pH, carbonate ion concentration, and saturation states of biologically important calcium carbonate minerals. These minerals are essential for the shells and skeletons of many marine organisms, and their depletion can have harmful effects, especially on coral reefs, shellfish, and plankton, which are the foundational components of the marine food web.

Pollution also significantly impacts the health of the oceans. Plastics, toxic runoff, and other contaminants harm marine life. Habitat destruction, from coastal erosion to the destruction of mangroves and seagrass beds, further weakens the ocean's ability to support the numerous species it houses. Invasive species introduced through shipping and other human activities are outcompeting native marine life, disrupting the delicate balance that has evolved over millennia.

The culmination of these pressures is a significant reduction in ocean fish stocks. This worrying trend threatens the marine food chain and endangers billions of people's livelihoods and food security worldwide. With over 40 percent of the ocean heavily impacted by human activities, no area remains unaffected by our influence.

## Data Analysis of Carbonate Chemistry


### H+ and HCO3- background

The ocean plays a crucial role in mitigating the effects of carbon dioxide (CO2) released into the atmosphere. It absorbs approximately 30% of the CO2 emissions resulting from human activities such as burning fossil fuels (e.g., car emissions) and changing land use (e.g., deforestation). However, as the levels of atmospheric CO2 increase, the amount of carbon dioxide absorbed by the ocean also increases. When CO2 dissolves in seawater, it leads to a series of chemical reactions that increase the concentration of hydrogen ions. This process has far-reaching implications for the ocean and the living organisms that inhabit it.

<img src="images/Ocean%20Acidification%20Infographic_800x350.png" alt="description" width="900" height="auto"/>

When carbon dioxide naturally dissolves into seawater, it forms carbonic acid (H2CO3) which dissociates into hydrogen ions (H+) and bicarbonate ions (HCO3-).

### Chemical Process Involved 

**CO<sub>2</sub> + H<sub>2</sub>O → H<sup>+</sup> + HCO<sub>3</sub><sup>-</sup>**




### Significance of the Equation

The equation illustrates the chemical process by which carbon dioxide dissolves into seawater, contributing to ocean acidification. This phenomenon has profound implications for marine ecosystems, including disruptions to coral reefs, changes in marine species behavior, and impacts on shell-forming organisms. Understanding these chemical dynamics is critical for assessing and addressing the consequences of ocean acidification on global marine biodiversity and ecosystems.

```python
import matplotlib.pyplot as plt
from scipy.stats import t
import numpy as np

# Assuming 'data_Hplus' and 'data_HCO3minus' contain the observations from the 'H+' and 'HCO3-' columns
data_Hplus = df['H+']
data_HCO3minus = df['HCO3-']

# Remove non-finite values (NaNs and infinities)
data_Hplus = data_Hplus[np.isfinite(data_Hplus)]
data_HCO3minus = data_HCO3minus[np.isfinite(data_HCO3minus)]

# MLE estimation for Student's t distribution
nu_mle_Hplus, mu_mle_studentt_Hplus, sigma_mle_studentt_Hplus = t.fit(data_Hplus)
nu_mle_HCO3minus, mu_mle_studentt_HCO3minus, sigma_mle_studentt_HCO3minus = t.fit(data_HCO3minus)

print("\nMLE Estimates for Student's t Distribution (H+):")
print("Degrees of Freedom (ν):", nu_mle_Hplus)
print("Mean (μ):", mu_mle_studentt_Hplus)
print("Standard Deviation (σ):", sigma_mle_studentt_Hplus)

print("\nMLE Estimates for Student's t Distribution (HCO3-):")
print("Degrees of Freedom (ν):", nu_mle_HCO3minus)
print("Mean (μ):", mu_mle_studentt_HCO3minus)
print("Standard Deviation (σ):", sigma_mle_studentt_HCO3minus)
```

```python
# Histogram and Student's t fit for H+
plt.figure(figsize=(10,6))
plt.hist(data_Hplus, bins=20, density=True, alpha=0.6, color='blue', label='Histogram (H+)')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = t.pdf(x, nu_mle_Hplus, mu_mle_studentt_Hplus, sigma_mle_studentt_Hplus)
plt.plot(x, p, 'k', linewidth=2, label="Student's t Fit")
plt.title("Histogram and Student's t Fit (H+)")
plt.grid(True)
plt.legend()
```

## Utilize Linear Regression and Polynomial Features to extract insights

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Define the data
x = df['H+'].values
y = df['HCO3-'].values

# Remove NaN values if present
valid_indices = ~np.isnan(x) & ~np.isnan(y)
x = x[valid_indices]
y = y[valid_indices]

# Create polynomial features
poly_features = PolynomialFeatures(degree=2)  # You can adjust the degree as needed
X_poly = poly_features.fit_transform(x[:, np.newaxis])

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict HCO3- levels for observed H+ values
observed_Hplus =  0.0091201083935591
X_observed = poly_features.transform(np.array([[observed_Hplus]]))
predicted_HCO3minus = model.predict(X_observed)

print("Predicted HCO3- level given observed H+ level of", observed_Hplus, ":", predicted_HCO3minus)

# Plot the relationship
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Observed Data')
plt.plot(x, model.predict(X_poly), color='red', label='Polynomial Regression')
plt.title('Relationship between H+ and HCO3-')
plt.xlabel('H+')
plt.ylabel('HCO3-')
plt.grid(True)
plt.legend()
plt.show()
```
